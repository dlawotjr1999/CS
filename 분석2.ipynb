```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

#multiple_ridge_regression
#default_alpha = 1.0
#alpha는 ridge, lasso, elasticnet의 파라미터

ridge = make_pipeline(PolynomialFeatures(degree=1),StandardScaler(),linear_model.Ridge())
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)

r2_ridge = r2_score(y_test, y_pred_ridge)
print(r2_ridge)
```

```python
#multiple_lasso_regression

#lasso = make_pipeline(PolynomialFeatures(degree=1),StandardScaler(),linear_model.Lasso())
lasso = linear_model.Lasso()
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)

r2_lasso = r2_score(y_test, y_pred_lasso)
print(r2_lasso)

c0 = lasso.coef_[0]
c1 = lasso.coef_[1]
c2 = lasso.coef_[2]
c3 = lasso.coef_[3]
c4 = lasso.coef_[4]
c5 = lasso.coef_[5]
b = lasso.intercept_

print(c0, c1, c2, c3, c4, c5, b)
```

```python
#elastic_net

elastic = make_pipeline(PolynomialFeatures(degree=3),StandardScaler(),linear_model.ElasticNet())
elastic.fit(X_train, y_train)
y_pred_elastic = elastic.predict(X_test)

r2_elastic = r2_score(y_test, y_pred_elastic)
print(r2_elastic)
```

```python
from sklearn.ensemble import GradientBoostingRegressor, GradientBoostingClassifier

#gbr

gbr = GradientBoostingRegressor(random_state=42).fit(X_train, y_train)
y_pred_gbr = gbr.predict(X_test)

r2_gbr = r2_score(y_test, y_pred_gbr)
print(r2_gbr)

```

```python
from google.colab import drive
drive. mount('/content/gdrive')

from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import OrthogonalMatchingPursuit
from sklearn.datasets import make_regression

df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/parking.csv", encoding = 'cp949')

#introduction

data_X = df[['대중교통', '사업체현황', '인구당자동차등록', '면적당인구수', '인구당생활유동인구','사업체현황']]
data_Y = df[['면적당주차장수']]

X_train, X_test, y_train, y_test = train_test_split(data_X,
data_Y, test_size=0.2, random_state=400)
```

```python

model = OrthogonalMatchingPursuit(n_nonzero_coefs=6)
model.fit(X_train, y_train)
```

```python
print("학습 데이터 점수: {}".format(model.score(X_train, y_train)))
print("평가 데이터 점수: {}".format(model.score(X_test, y_test)))
```
