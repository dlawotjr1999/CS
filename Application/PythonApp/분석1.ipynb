```python
from google.colab import drive
drive. mount('/content/gdrive')

from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/전처리ㅇㅇ.csv", encoding = 'cp949')
weights=[];

X1 = df["면적"]

y = df["주차장수"]

regr1 = linear_model.LinearRegression()
regr1.fit(X1.values.reshape(-1,1), y)

score_A=[];
score_B=[];
score_C=[];
score_D=[];
score_E=[];
score_F=[];


difference_A=[];
difference_B=[];
difference_C=[];
difference_D=[];
difference_E=[];
difference_F=[];

#plt.figure(figsize=(120,60))
plt.plot(X1, y, 'o')
plt.xlabel("Area")
plt.ylabel("Parking Space")
plt.plot(X1, regr1.predict(X1.values.reshape(-1,1)))

for i in range(len(df)):
    row = df.iloc[i]
    X = row['면적']
    y = row['주차장수']

    plt.text(X, y, i)

#plt.xticks([i for i in range(0, 1500, 100)])
plt.yticks([i for i in range(0, 60000, 5000)])
plt.show()


print("Slope & Intercept of LinearRegression :", regr1.coef_,regr1.intercept_)
print('\n')

x1 = df["면적"]
y_res = df["주차장수"]

y1 = regr1.coef_*x1 + regr1.intercept_

y1_diff = y1 - y_res

for i in range (0,424):
  difference_A.append(round((float(y1_diff[i]/y1[i])*100),2))

print(y1_diff[0])
```

```python
from google.colab import drive
drive. mount('/content/gdrive')

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

diff = pd.read_csv("/content/gdrive/MyDrive/Colab Notebooks/Diff_data.csv", encoding = 'cp949')

Cluster_X = diff[['면적', '인구수', '생활유동인구', '자동차등록', '대중교통']]
Cluster_Y = diff[['주차장수']]

# Cluster_X 정규화(일단은 넣음 필요하면 갖다쓰기)
sc = StandardScaler()
cc_scaled = sc.fit_transform(Cluster_X)
pd.DataFrame(cc_scaled)

ks = range(1,20)
inertias = []
for k in ks :
  model = KMeans(n_clusters = k, n_init = 5)
  model.fit(cc_scaled)
  inertias.append(model.inertia_)

plt.figure(figsize = (15,6))
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()

#k = 11 이용
```

```python
#k = 11로 clustering
kmeans = KMeans(n_clusters=11, init = 'k-means++')
clusters = kmeans.fit(cc_scaled)

Cluster_X['cluster'] = clusters.labels_

# 시각화를 위한 차원 축소
X = cc_scaled.copy()
pca = PCA(n_components=2)
pca.fit(X)
x_pca = pca.transform(X)
pca_df = pd.DataFrame(x_pca)
pca_df['cluster'] = Cluster_X['cluster']

#시각화
axs = plt.subplots()
axs = sns.scatterplot(0,1,hue='cluster', data=pca_df)

Cluster_X.insert(0,'동명', diff['동명'])
Cluster_X.insert(6, '주차장수', diff['주차장수'])

Cluster_X




```
